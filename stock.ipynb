# 1. Price

```python
from vnstock import Quote
quote = Quote(symbol='TCB', source='VCI')
```

```python
import matplotlib.pyplot as plt
```

## History

```python
df = quote.history(start='2025-01-01', end='2025-09-05', interval='1D')
plt.figure(figsize=(20,6))
plt.plot(df['time'], df['close'], label='TCB')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('TCB Close Price Over Time')
plt.legend()
plt.show()
```

# Intraday

```python
# D liu khp lnh trong ngy giao dch realtime hoc ngy gn nht (ngoi gi giao dch)
quote.intraday(symbol='TCB', page_size=10_000, show_log=False)
```

```python
# Bc gi v khi lng giao dch: realtime
quote.price_depth('TCB')
```

# VN Index

```python
quote = Quote(symbol='VNINDEX', source='VCI')
df_VNI = quote.history(symbol='VNINDEX', start='2025-01-01', end='2025-09-05', interval='1D')
plt.figure(figsize=(24,6))
plt.plot(df_VNI['time'], df_VNI['close'], label='VNINDEX')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('VN Index Close Price Over Time')
plt.legend()
plt.show()
```
